```python
import json


with open("data/processed/nat/semantic_dimension/semantic_dimension_binary_gt.json", "r") as f:
    data = json.load(f)
```

```python
rare = []
common = []
for lang in data.keys():
    lang_item_list = data[lang]
    for item in lang_item_list:
        word = item["word"]
        ipa = item["ipa"]
        language = item["language"]
        word_group = item["word_group"]
        for dimension, value in item["dimensions"].items():
            answer = value["answer"]
            
            element = {
                "word": word,
                "ipa": ipa,
                "language": language,
                "word_group": word_group,
                "answer": answer,
                "dimension": dimension,
            }
            
            if word_group == "rare":
                rare.append(element)
            elif word_group == "common":
                common.append(element)
            else:
                raise ValueError(f"Invalid word group: {word_group}")
            

print(len(rare))
print(len(common))
```

```python
# save the data to a json file

import json


with open("rare.json", "w") as f:
    json.dump(rare, f)

with open("common.json", "w") as f:
    json.dump(common, f)
```
